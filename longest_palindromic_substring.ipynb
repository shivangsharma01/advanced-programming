```python
class Solution:
    def logest_palindromic_substring(self, s: str)-> str:

        n = len(s)
        ans = ""

        # palindrome can be odd/even length
        # palindrome always has a centre
        # we iterate through the center of the palindrome to the outside of it while both the outside characters are same return s[l+1:r]


        # palidrome mirrors around its centre - so we iterate around the centre and expand outwards
        # first thing to do is write the expand()

        def expand(l,r):

            while (l>0 and r<n and s[l]==s[r]):
                l-=1
                r+=1

            return s[l+1:r]
        
        for i in range(n):

            odd = expand(i,i)
            even = expand(i+1,i)

            if len(odd) > len(ans):
                ans = odd
            if len(even)> len(ans):
                ans = even
        return ans
```
