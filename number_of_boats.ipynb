## Find the min number of boats leetcode 881

### Brute force is :

find all the possible pairs and return the pair with min length

But this will obviously consume TC of O(n^2)

More optimised approach will be to sort the array - i will get the heaviest and the lightest people in the array - leftmost is the lightest and the righmost is the lightest

Greedy approach
+
2 pointer




```python
class Solution:

    def numRescueBoats(self, people: list[int], limit:int)->int:

        people.sort() # follow a greedy approach

        left = 0
        right = len(people)-1

        boats = 0

        while left<=right: # because here even with 2 pointer approach - the last person will require a boat

            # depart the heaviest person in the first boat - if we can pair him up with a lighweight - GREAT so right pointer will definitely decrease in every loop
            # left only increases when limit can fit the people[right]+people[left]

            if people[left]+people[right]<=limit:
                left +=1

            right-=1

            boats+=1

        return boats
```
