# add two numbers

```python
# given are 2 linked lists- add them and reverse the list and return the output
class ListNode:
    def __init__(self,val=0,next =None):
        self.val = val
        self.next = next

class Solution:
    def add_two_nums(self, l1: Optional[ListNode],l2: Optional[ListNode]):

        # dummy node needs to be created so that we can point the new linkedlist to dummy.next
        dummy = ListNode(0)

        # make a current node after which we will keep on adding the new nodes and later we will return curr.next
        curr = dummy.next

        # l1 l2 and carry are important
        carry=0

        while l1 or l2 or carry:

            if l1:
                x = l1.val
            else:
                x =0 
            if l2:
                y = l2.val
            else:
                x=0

            new = x+y+carry

            digit = new%10
            carry = new//10

            curr.next = ListNode(digit)
            curr = curr.next

            if l1:
                l1=l1.next
            if l2:
                l2=l2.next

        return dummy.next
```
