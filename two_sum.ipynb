```python
#brute
def two_sum(nums, target):
    n= len(nums)

    for i in range(n):
        for j in range(i+1,n):

            if target == nums[i]+nums[j]:
                return [i,j]
```

```python
#optimised
def two_sum(nums, target):
    
    seen = {} # hashmap to reduce the TC

    for i,num in enumerate(nums):

        diff = target - num

        if diff in seen:
            return [seen[diff], i]
        
        seen[num] = i
```
