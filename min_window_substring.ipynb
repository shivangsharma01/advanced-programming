```python
# Simple steps to remeber the problem

# Mnuemonic - going for shopping and having a shopping list

# Create a need hashmap and check usin counter
# Create a window hashmap
# Initialise the res vars
# Iterate through s using right pointer
# keep adding the chars in the hashmap
# if requirement satisfied -  have+=1
# while have == need_count:
#     update answer
#     remove left
#     if requirement not satisfied:
#         have -=1
#     left+=1
```

```python
from collections import Counter

class Solution:
    def minWindow(self, s: str, t: str) -> str:
        
        if len(s)<len(t):
            return ""

        have=0
        left = 0
        need = Counter(t)

        window = {}

        res_len = float('inf')
        res_left = 0
        res_right = 0

        for right in range(len(s)):

            char = s[right]

            window[char] = window.get(char,0)+1

            if char in need and window[char]==need[char]:
                have+=1
            
            while have == len(need):

                if res_len> right-left+1:

                    res_len = right-left+1
                    res_left = left
                    res_right = right

                window[s[left]]-=1

                if s[left] in need and window[s[left]]<need[s[left]]:
                    have-=1
                
                left+=1
        
        if res_len==float('inf'):
            return ""
        
        return s[res_left:res_right+1]
```
