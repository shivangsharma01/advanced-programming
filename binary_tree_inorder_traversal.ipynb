#### USE RECUSRION TO ITERATE THROUGH THE TREE AND THEN YOU CAN APPLY PREORDER INORDER OR POSTORDER

```python
from typing import Optional

class TreeNode:
    def __init__(self, val = 0, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def inorder_traversal(self,root: Optional[TreeNode])->list[int]: # Optional says-  input can either be None or an object of TreeNode class
        
        if root is None:
            return None
        
        ans = []

        def inorder(node):

            # inorder - left root right

            inorder(node.left)

            ans.append(node.val)

            inorder(node.right)

        return ans
```

## Recursive solution is trivial - SOLVE IT ITERTIVELY

```python
class Solution:
    def inorder_traversal(self,root: Optional[TreeNode])->list[int]: # Optional says -  input can either be None or an object of TreeNode class
        

        stack = [] # even the recursive soln maintains a stack for the nodes it has traversed

        ans= []

        curr = root

        while curr or stack: # is there a node i have not iterated yet? - is there a node in the stack that i have not processed yet?


            # inorder: LEFT - NODE - RIGHT
            while curr: # completely go through the left side of the tree
                stack.append(curr)
                curr = curr.left
            
            # process the current node

            curr = stack.pop()
            ans.append(curr.val)

            # process the rightmost node
            curr = curr.right

        return ans 
```
