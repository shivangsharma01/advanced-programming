# understand the binary pattern

```python
piles=[1,4,5,6,5,3,2]
h=8

def can_finish(speed):

    hours = 0

    for pile in piles:

        # hours = ciel(pile/speed)
        hours = (pile+speed-1)//speed

    return hours >= h
```

```python
class Solution:
    def koko(self, piles:list[int], h:int)->int:

        l=0
        r = max(piles)
        
        def can_finish(speed):

            hours = 0

            for pile in piles:

                # hours+= math.ceil(pile/speed)
                hours += (pile+speed-1)//speed

            return hours>=h
        
        while l<r:

            mid = (l+r)//2

            if can_finish(mid):
                # mid satisfies the condition
                r = mid
            else:
                # too slow
                l= mid+1
        return l
```
