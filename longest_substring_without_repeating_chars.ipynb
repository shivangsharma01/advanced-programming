# use a sliding window for O(n) operation and use a set to store the unique characters

```python
class Solution:
    def lswrc(self, s: str) -> int:

        window = set()

        l = 0
        ans =0 

        for r in range(len(s)):

            while s[r] in window:

                window.remove(s[l])
                l+=1
            ans = max(ans, r-l+1)
        print(ans)
```

# hashmap approach

```python
class Solution:
    def lswrc(self, s:int)->int:

        last_seen = {}

        l = 0
        ans = 0

        for r in range(len(s)):

            if s[r] in last_seen:

                l = max(l, last_seen[s[r]]+1)

            last_seen[s[r]] =r

            ans = max(ans, r-l+1)
        return ans
```
