```python
# this is a boundary problem

# we can do it using an array search also but it will take O(N) time

# lets build the reusable block

def lower_bound(nums, target)->int:
    l,r = 0, len(nums)

    while l<r:
        mid = (l+r)//2

        if nums[mid]>target: # upper bound
            r = mid
        else:
            l = mid+1 # lower bound nums[mid] <= target

    return l # returns the first index where nums[i] >= target

first_index = lower_bound(nums, target)

if first_index == len(nums) or nums[first_index]!=target:
    print([-1,-1])

second_index = lower_bound(nums, target+1) -1
```
