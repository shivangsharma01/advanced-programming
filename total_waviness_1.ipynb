```python
class Solution:
    def total_waviness(self, num1 :int, num2: int):

        def waviness(num):

            digits = str(num) # convert the number to string version e.g.: 120 -> ['1','2','0']
            answer = 0

            for i in range(1, len(digits)-1):

                left = digits[i-1]
                curr = digits[i]
                right = digits[i+1]

                # check the peak and waviness
                if curr<left and curr<right: # valley
                    answer+=1
                if curr>left and curr>right: # peak
                    answer+=1

            return answer
        ans = 0

        for num in range(num1, num2+1): # because python's range function excludes the last value
            ans+=waviness(num)
        print(waviness)
```

#### Time Complexity: O((num2 − num1 + 1) × D), where D is the number of digits in the largest number.
#### Space Complexity: O(D) due to the temporary string representation of each number. If the digit storage is not counted as auxiliary space, then the auxiliary space is O(1).
