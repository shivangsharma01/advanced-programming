```python
class Solution:
    def clock_angle(self, hours:int, minutes:int)->int:

        hour %=12 # because if the hour==12 then the angle will be 360 degree which is not valid

        # hour angle will contain the hour (360/12=30 multiple of hour) i.e. 30*hours and in 60 minutes the hour hand moves 30 degree (30/60 = 0.5 degree) every minute 
        # so 30*hours + 0.5*minutes
        
        hour_angle = 30*hours + 0.5*minutes


        # the minute hand moves 360 degree in 60 minutes 360/60 = 6 - hence at any minute the angle of the minute hand is a multiple of 6
        minutes_angle = 6*minutes

        diff = abs(hour_angle - minutes_angle)

        return min(diff, 360-diff)
```
