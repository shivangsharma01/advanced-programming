```python
class Solution:
    def longest_subsequence_length(self, nums:list[int])->int:
        # i will create a set as it removes the possibility of duplicates and gives me a constant time lookup
        num_set = set(nums)

        longest = 0

        for num in num_set:

            if num-1 not in num_set:

                current = num
                length = 1

                while current+1 in num_set:

                    current+=1
                    length+=1
            
                longest = max(longest,length)

        return longest
```
