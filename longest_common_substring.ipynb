# Longest common substring

Write a function to find the longest common prefix string amongst an array of strings.

If there is no common prefix, return an empty string "".

Example:

Input: strs = ["flower","flow","flight"]
Output: "fl"

#### Lets say that i have to do this problem manually and solve it myself - how will i do it?
#### I will compare first characters of all strings - if same
#### compare the second characters of all the strings - if same move to next and keep appending in prefix string

#### where will this fail - one string is smaller and ends early - we can add if i==len(str) then return the current prefix
#### other can be that we dont find the right char - both cases will return the currrent prefix

#### I will treat all the strings as rows in a table and then compare thir numbers 1-by-1

```python
# rebuilding the intuition

class Solution:

    def lcs(self, strs:list[str])->str:

        if len(strs)<0:
            return ""
        
        prefix = ""

        for i in range(strs[0]):

            char = str[0][i]

            for word in str[1:]:

                if i==len(word) or char != word[i]:
                    return prefix
                
            prefix+=char
        return prefix
```
