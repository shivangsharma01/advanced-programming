```python
from collections import defaultdict, deque

class Solution:
    def can_finish(self, numCourses: int, prerequisites: list[list[int]]) -> bool:
        
        graph = defaultdict(list) # helps with indexing as compared to a dict(list)

        # create an indegree which will give you list of each course and how many prerequisites it requires
        # this helps me understand that if any course has 0 prereq i can do it right away

        indegree = [0]*numCourses

        for course, prereq in prerequisites:

            graph[prereq].append(course) # eg if we have (0,1) graph will be {1:[0]} then if we get (2,1), graph: {1:[0,2]}
            indegree+=1

        queue = deque() # why only a deque - O(1) operations and i can popleft and popright wihtout sorting

        for course in range(numCourses):
            if indegree[course]==0:
                queue.append(course)
        
        completed = 0
        
        while queue:

            course = queue.popleft() # explains why we are using a deque

            # once i have done the course - add it to the completed courses - also check the list for other courses
            completed +=1

            for neighbor in graph[course]:
                indegree[neighbor]-=1

                if indegree[neighbor]==0:
                    queue.append(neighbor)
        return completed == numCourses
```
