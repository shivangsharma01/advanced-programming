```python
class Solution:
    def totalFruit(self, fruits: list[int]) -> int:
        
        # this is a classical sliding window problem 
        # find the most optimal window
        # i can pick only 1 fruit from 1 tree
        # i only have 2 baskets - my dict's len cannot exceed 2 here
        # once it exceeds 2 - i will remove the leftmost number in the hashmap

        ans = 0
        left = 0
        
        count = dict()


        for right in range(len(fruits)):

            # Now the hashmap will contain this fruit and the value will be added by 1
            # if the value does not exist i will consider it as 0 and give it initial value 1

            count[right] = count.get(fruits[right],0) + 1

            if len(count)>2:

                count[left] -=1
                
                if count[left]==0:
                    del count[left]

                # i have managed the left pointer - now need to fix ans

                ans = max(ans,left-right+1)
        
        return ans
```

```python
class Solution:
    def totalFruit(self, fruits: list[int]) -> int:

        ans = 0
        left = 0
        count = {}

        for right in range(len(fruits)):

            fruit = fruits[right]

            count[fruit] = count.get(fruit,0)+1 # Hashmap O(1) Lookup

            while len(count)>2:

                left_fruit = fruits[left]

                count[left_fruit]-=1

                if count[left_fruit]==0:
                    del count[left_fruit]
                
                left+=1

            ans = max(ans, right-left+1)
        return ans
```

# 3rd iteration

```python
class Solution:
    def totalFruit(self, fruits: List[int]) -> int:
        
        # sliding window

        ans = 0
        left = 0
        count = dict() # hashmap

        for right in range(len(fruits)):

            fruit = fruits[right]

            # simple hashmap with fruits and their counts
            count[fruit] = count.get(fruit,0) +1

            while len(count)>2:

                left_fruit = fruits[left]
                count[left_fruit]-=1

                if count[left_fruit]==0:
                    del count[left_fruit]
                
                # once we delete the left pointer we must move the left pointer ahead by 1
                left+=1
                
            ans = max(ans, right-left+1)
        return ans
```
