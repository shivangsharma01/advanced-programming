## Brute force

#### most water = width*min_height(left,right) - Brute force - for every i go check every j for it - calculate the water in the container - return max value

## Optimised Aproach

```python
# container with the max amount of water will definitely have very high wall be it left or right

# keep pointers and move the pointers such that we find two walls and keep the tall wall fixed and keep moving the shorter value

class Solution:
    def max_water_in_container(self, nums:list[int])->int:

        left= 0 
        right = len(nums)-1
        maxarea = 0

        while left<right:

            width = nums[right] - nums[left] 

            minheight = min(nums[left], nums[right])

            area = width*minheight

            max = max(maxarea,area)

            if nums[left]<nums[right]:
                left+=1
            else:
                right-=1
        
        print(maxarea)
```
