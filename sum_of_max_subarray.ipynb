```python
class Solution:
    def max_subarray(self,nums):

        curr = nums[0]
        best = nums[0]

        for i in range(1, nums):

            # Use kadane algo here - at every index check - is the current index helping me - if not re-initialise the current var

            curr = max(nums[i], curr+nums[i]) # ONLY compare the num at the current index with (addition of curr with it)
            best = max(best, curr)

        return best
```
