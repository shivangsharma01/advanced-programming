# prefix/suffix array - pointers (optimal)

```python
class Solution:
    def trap(self, height: list[int]) -> int:
        
        # prefix-suffix array - to pointer based optimised approach


        # what i will do is for each height[i] i look at- i will find it's left max and right max and then take the x=min(leftmax,rightmax) and height[i] - x -  this will give me how much water is present there and i will add it in the total value of the water

        # but wait - for every height[i] if i have to compute leftmax and rightmax - it is very expensive - O(n^2)

        # so i will store the values in left and right in an array and take the max(left) and max(right) for every height[i] and then find the trapped water but this is also expensive

        # better to store leftmax and rightmax pointer and move them only if there is any value larger than them

        left = 0
        right = len(height)-1
        leftmax = 0
        rightmax = 0

        water = 0

        while left<right:

            if height[left] < height[right]: # this means that right is not a problem and it will help us hold the water - left is the problem and we must find leftmax
                if height[left]>=leftmax:

                    leftmax =height[left]
                else:
                    water+= leftmax - height[left]
                
                left+=1
            else:
                # if height[left]>= height[right]
                # if right is the problem we must find rightmax

                if height[right]>= rightmax:
                    rightmax = height[right]
                else:
                    water+= rightmax - height[right]
                right-=1
        return water
```
