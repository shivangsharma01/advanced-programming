# Is a given integer palindrome or not?

```python
class Solution:
    def palindrome_number(self, x: int)->bool:

        return str(x)==str(x)[::-1] # Convert int to string and compare with the inversion
```

# Can you solve without strings

```python
class Solution:
    def palindrome_number(self, x: int)->bool:

        # reverse the number and create a rev variable - and then match and return bool

        if x<0 or (x%10==0 and x!=0): # "-121" is not a palindrome OR if the number is single digit - NOT A PALINDROME
            return False # edge cases have been taken out
        
        rev = 0 # reversing the integer and creating a copy

        while x>rev: # if x exceeds rev then it is void

            digit = x%10 # digit at the tens place

            rev = rev*10+digit # initialise rev as 0 and then keep adding up from the back as we iterate

            x = x//10 # reduce one index position ever time
        
        return  x == rev or x == rev//10
    
# Solution goes from 0123456789 - len = 10 so tc -> O(log10n) or O(LogN) : SC is Constant
```
