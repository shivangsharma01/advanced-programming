# Max frequency after subarray optimisation 

```python
class Solution:
    def max_freq(self, nums: list[int], k:int) -> int:

        base = nums.count(k)
        ans = base
        
        # find how many  
        for v in set(nums):
            
            if v==k:
                continue
            
            curr = 0
            best_gain = 0

            for num in nums:

                if num==v:
                    score =1
                elif num == k:
                    score =-1
                else:
                    score =0
                
                # kadane
                curr = max(num, num+score)
                best_gain = max(best_gain, curr)

            ans = max(ans, base + best_gain)
        print(ans)
```
