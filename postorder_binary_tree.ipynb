```python
from typing import Optional

class TreeNode:
    def __init__(self, val = 0, left = None, right = None):
        self.val = val
        self.right = right
        self.left = left

class Solution:
    def postorder(self, root: Optional[TreeNode])-> list[int]:

        ans = []

        def postorder(node):

            if node is None:
                return None

            postorder(left)
            postorder(right)

            ans.append(node.val)

        return ans
```
