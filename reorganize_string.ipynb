```python
s= "aaabc"

# dict
freq = {
    "a":"3",
    "b":"2",
    "c":"1"
}
#resulting string
res =""

# I would have to pop and adjust the frquency if i go with a greedy approach here
# Insertion and sorting will be very expensive here

from collections import Counter

freq = Counter(s) # gives me the character and the frequency of it


import heapq

# declaring a max heap
heap = []

for ch, count in freq.items():

    heapq.heappush(heap, (-count,ch))

if max(freq.values()) > (len(s)+1)//2:
    print("")

result = []

prev_count = 0
prev_ch = ""

while heap:

    count , ch = heapq.heappop(heap)

    if count < 0:
        heapq.heappush(heap, (prev_count, prev_ch))

    result.append(ch)
    count+=1

    prev_ch = ch
    prev_count = count
    
print("".join(result))
```
